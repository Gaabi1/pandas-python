Limpeza de Dados

```python
df = pd.read_csv('funcionarios_nulos.csv')
df
```

```python
df.dropna()
```

```python
#Remover só as linha que tem valores nulos pra coluna idade
df.dropna(subset="Idade")
```

```python
#Preencher os valores nulos com outros valores
media_valores= df['Idade'].mean()
df['Idade'] = df['Idade'].fillna(media_valores)
df
```

```python
#Trocar tipos de Dados
df['Idade'] = df['Idade'].astype(int)
df
```

Contando Ocorrências

```python
df = pd.read_csv('funcionarios.csv')
```

```python
#Conta funcionarios ativos
df['Ativo'].value_counts()
```

```python
#Numero de funcionarios em cada departamento
df['ID Departamento'].value_counts()
```

Operações de Agregação

```python
df.groupby('ID Departamento')
```

```python
#Média salária por departamento
df.groupby('ID Departamento')['Salário'].mean()
```

```python
df.groupby('ID Departamento')['Salário'].sum()
```

```python
df.groupby('ID Departamento')['Salário'].max()
```

Ordenando Dados

```python
#Do menor pro maior
df.sort_values(by='Salário')
```

```python
#Do maior pro menor
df.sort_values(by='Salário', ascending=False)
```

Combinando Dados

```python
df_depart = pd.read_csv('departamentos.csv')
df_depart
```

```python
df_final = pd.merge(df, df_depart)
df_final
```

```python
#Especificar caso as duas tabelas não fossem iguais
df_final = pd.merge(df, df_depart, left_on='ID Departamento', right_on='ID Departamento')
df_final
```

Empilhando

```python
df_novos= pd.read_csv('novos_funcionarios.csv')
df_novos
```

```python
#Uni tudo verticalmente e recria id do zero
df_todos= pd.concat([df, df_novos], ignore_index=True)
df_todos
```

Escrevendo dados em arquivo

```python
df_todos.to_csv('funcionarios_final.csv')
```
